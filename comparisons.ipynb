## Comparison of PSO and GA

```python
from optproblems import cec2005
import ga.ga_functions as ga
import numpy as np
```

### Comparing Different Benchmark Functions for PSO and GA

```python
function0 = cec2005.F1
function1 = cec2005.F2
function2 = cec2005.F9
function3 = cec2005.F11

test_f0 = ga.GA(function0(10), -100, 100, 10, 50, 100) # -450 minima
test_f1 = ga.GA(function1(10), -100, 100, 10, 50, 400) # -450 minima
test_f2 = ga.GA(function2(10), -5, 5, 10, 50, 100) # -330 minima
test_f3 = ga.GA(function3(10), -0.5, 0.5, 10, 50, 1100) # 90 minima

variablef0, minimaf0, genf0 = test_f0.run_ga()
variablef1, minimaf1, genf1 = test_f1.run_ga()
variablef2, minimaf2, genf2 = test_f2.run_ga()
variablef3, minimaf3, genf3 = test_f3.run_ga()

print(f"\nThe minima reached for function cec2005.F1() is {minimaf0} at Generation {genf0}.\nBest population is {variablef0}")
print(f"\nThe minima reached for function cec2005.F2() is {minimaf1} at Generation {genf1}.\nBest population is {variablef1}")
print(f"\nThe minima reached for function cec2005.F9() is {minimaf2} at Generation {genf2}.\nBest population is {variablef2}")
print(f"\nThe minima reached for function cec2005.F11() is {minimaf3} at Generation {genf3}.\nBest population is {variablef3}")
```

```python
import pso.PSO as PSO
import pso.particle 

test_f0 = PSO.PSO(lbound = -100, ubound = 100, size = 50, iters =100, obj=function0) # -450 minima
test_f1 = PSO.PSO(lbound = -100, ubound = 100, size = 50, iters = 100, obj=function1) # -450 minima
test_f2 = PSO.PSO(lbound = -5, ubound = 5, size =50, iters = 100, obj=function2) # -330 minima
test_f3 = PSO.PSO(lbound = -0.5, ubound = 0.5, size = 50, iters = 100, obj=function3) # 90 minima

test_f0.optimize()
test_f1.optimize()
test_f2.optimize() 
test_f3.optimize()

print(f"\nThe minima reached for function cec2005.F1() is {test_f0.gBest} at Iteration {test_f0.bestiter}.\nThe best position is {test_f0.gBestPos}")
print(f"\nThe minima reached for function cec2005.F2() is {test_f1.gBest} at Iteration {test_f1.bestiter}.\nThe best position is {test_f1.gBestPos}")
print(f"\nThe minima reached for function cec2005.F9() is {test_f2.gBest} at Iteration {test_f2.bestiter}.\nThe best position is {test_f2.gBestPos}")
print(f"\nThe minima reached for function cec2005.F11() is {test_f3.gBest} at Iteration {test_f2.bestiter}.\nThe best position is {test_f3.gBestPos}")
```
